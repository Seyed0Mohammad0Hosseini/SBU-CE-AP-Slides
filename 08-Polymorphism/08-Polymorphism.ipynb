  <h3>بازنویسی متدهای ابرکلاس در زیرکلاس‌ها</h3>


  <p>
همان‌طور که در اسلاید قبل گفتیم، یک زیرکلاس متدهای پدرش یا همان ابرکلاس را ارث‌بری می‌کند. بعضی وقت‌ها این زیرکلاس نیاز دارد پیاده‌سازی یک متد تعریف‌شده در ابرکلاسش را تغییر دهد. به این کار به اصطلاح<strong> بازنویسی متد</strong> یا <strong> method overriding </strong> می‌گوییم.
  </p>

<p>
به عبارت دیگر، هرگاه متدی در زیرکلاس بخواهد پیاده‌سازی خاص خود از متد ارث برده‌شده از ابرکلاس را داشته باشد، از بازنویسی استفاده می‌کند.
</p>

  <h3>نکات بازنویسی متد</h3>
  <p>
  <strong>چرا یک متد استاتیک قابل بازنویسی نمی‌باشد؟</strong>
  </p>
  <p>
چون متد استاتیک به یک کلاس مقید شده است نه به اشیای کلاس. اما یک متد نمونه به یک شی از کلاس مقید شده است. متدها و فیلدهای استاتیک متعلق به تعریف کلاس در حافظه ایستا هستند که در زمان اجرا قابل تغییر نمی‌باشندد. اما متدهای نمونه در حافظه heap ایجاد می‌شوند و امکان بازنویسی یا تغییر آنها وجود دارد.
  </p>
  <p>
  
  </p>
  <p>
  <strong> آیا می‌توانیم متد main در جاوا را بازنویسی کنیم؟</strong>
  </p>
  <p>
جواب مشخصا خیر است چون متد main به صورت استاتیک تعریف می‌شود.
  </p>

  <p>
  <strong> نکته:</strong> یک متد نمونه تنها در صورت در دسترس بودن، قابل بازنویسی می‌باشد. بنابراین یک متد private نمی‌تواند بازنویسی شود، چون در خارج از کلاس خود قابل دسترسی نمی‌باشد. اگر  متد تعریف شده در یک زیرکلاس همنام با متدی در ابرکلاسش باشد که private تعریف شده است، این دو متد کاملا بی‌ربط با یکدیگر در نظر گرفته می‌شوند.
مانند متد نمونه، یک متد استاتیک نیز می‌تواند ارث بری شود.
  </p>
<p>
 با این حال، یک متد استاتیک نمی‌تواند بازنویسی شود. اگر متد استاتیک تعریف شده در یک ابرکلاس در یک زیرکلاسش بازتعریف شود، متد تعریف شده در ابرکلاس پنهان(پوشیده) خواهد شد.
</p>

  <h3>overloading در مقایسه با overriding</h3>

  <p>
  <strong>سه تفاوت اصلی</strong> میان بازنویسی و سربارگذاری متدها وجود دارد: 
  </p>
  <p> بازنویسی متد به برنامه‌نویس امکان می‌دهد پیاده‌سازی خاص متد ارث بری‌شده از کلاس والد را ایجاد کند و همواره در دو کلاسی انجام می‌شود که رابطه ارث‌ بری میان آنها وجود دارد. و لازم به ذکر است که پارامترها باید یکسان باشد</p>
  <p>
  ولی سربارگذاری متد به هدف افزایش خوانایی برنامه‌ها ایجاد می‌شود و درون یک کلاس انجام می‌شود و برخلاف بازنویسی، پارامترها باید متفاوت باشند.
  </p>


```python
//Overriding

public class Test {
    public static void main(String[] args) {
        Aa a = new A();
        a.p(10);
        a.p(10.0);
    }
}

class B {
    public void p(double i) {
        System.out.println(i * 2);
    }
}

class A extends B {
    //This method overrides the method in B
    public void p(double i) {
        System.out.println(i);
    }
}
```

```python
//Overloading

public class Test {
    public static void main(String[] args) {
        Aa a = new A();
        a.p(10);
        a.p(10.0);
    }
}

class B {
    public void p(double i) {
        System.out.println(i * 2);
    }
}

class A extends B {
    //This method overloads the method in B
    public void p(int i) {
        System.out.println(i);
    }
}
```

<h3>مفهوم پلی مورفیسم </h3>
<p>
پلی مورفیسم زمان اجرا، فرآیندی است که فراخوانی یک متد بازنویسی شده در زمان اجرا و نه در زمان کامپایل انجام می‌شود. در این فرآیند، یک متد بازنویسی شده از طریق یک متغیر ارجاعی از کلاس والد (ریموت کنترل کلاس والد) فراخوانی می‌شود. تعیین اینکه کدام متد فراخوانی شود، به شی‌ای بستگی دارد که متغیر ارجاعی به عنوان آرگومان به آن اشاره می‌کند.

<h3>مثالی از پلی مورفیسم زمان اجرا</h3>
<p>
در این مثال دو کلاس Bike و Splender تعریف شده است، کلاس Splender از کلاس Bike ارث بری می‌کند و متد ()run را بازنویسی می‌کند.
</p>
<p>
اینک، متد ()run را از طریق متغیر ارجاعی کلاس والد فراخوانی می‌کنیم.
</p>
<p>
چون متغیر ارجاعی به شیی زیرکلاس اشاره می‌کند و متد زیرکلاس، متد ابرکلاسش را بازنویسی کرده است، در زمان اجرا متد زیرکلاسش فراخوانی می‌شود.
</p>
<p>
چون فراخوانی متد توسط JVM و نه کامپایلر انجام می‌شود، این فراخوانی به پلی مورفیسم زمان اجرا شهرت دارد.
</p>

```python
class Bike{
    void run(){
        System.out.println("running");
    }
}
class Splender extends Bike{
    void run(){
        System.out.println("running safely with 60km");
    }

    public static void main(String[] args){
        Bike b = new Splender(); //upcasting
        b.run();
    }
}
```

<h3>پلی مورفیسم زمان اجرا با فیلد داده‌ای</h3>
<p>
همیشه متدها هستند که بازنویسی می‌شوند و نه فیلدهای داده‌ای کلاس.
</p>
<p>
بنابراین پلی مورفیسم بر روی فیلدهای داده‌ای قابل دستیابی نخواهد بود. 
</p>
<p>
در مثال زیر، هر دو کلاس دارای فیلد داده‌ای speedlimit هستند. دسترسی به این فیلد داده‌ای توسط متغیر ارجاعی کلاس والد است که به  شی زیرکلاس اشاره می‌کند یا به عبارتی ریموت کنترل آن است.

<p>
چون ما به فیلد داده‌ای دسترسی داریم که بازنویسی نشده است و در اصل امکان بازنویسی آن وجود ندارد، همیشه فقط مقدار فیلد داده‌ای کلاس والد قابل دسترسی می‌باشد.
</p>

اشیا دارای نوع هستند: یک شی، نمونه‌ای از یک کلاس مشخص جاوا می‌باشد، اما در عین حال نمونه‌ای از کلاس والدش هم هست.
</p>
<p style="text-align: left;"><code>{}class Animal</code></p>
<p style="text-align: left;"><code>}class Dog extends Animal</code></p>
<p style="text-align: left;"><code>public static void main(String[] args){ Dog d1 = new Dog() }</code></p>
<p style="text-align: left;"><code>{</code></p>
<hr class="divider" />
<p>
<strong>مقیدسازی ایستا:</strong> هرگاه نوع یک شی در زمان کامپایل (توسط کامپایلر) مشخص شود، می‌گوییم <strong>مقیدسازی ایستا</strong> انجام شده است.
</p>
<hr class="divider" />
<p>
<strong>مقیدسازی پویا:</strong> هرگاه نوع شی در زمان اجرا مشخص شود، می‌گوییم <strong>مقیدسازی پویا</strong> انجام شده است.
</p>
<p>
مقیدسازی پویا به این شکل عمل می‌کند: فرض کنید شی o یک نمونه از کلاس‌های Cn ,Cn-1 , ...,C2 ,C1 باشد، به طوری که C1 یک زیرکلاس C2 و C2 یک زیرکلاس C3 و … و Cn-1 یک کلاس Cn باشد. یعنی Cn عمومی‌ترین کلاس باشد و C1 اختصاصی‌ترین کلاس باشد. در جاوا Cn همان کلاس Object است. اگر o یک متد p را فراخوانی کند، JVM به دنبال پیاده‌سازی متد p به ترتیب در C1, C2 , … , Cn می‌گردد تا زمانی که متد مربوطه را بیابد. به محض یافتن متد، جستجو خاتمه یافته و پیاده‌سازی شناسایی شده فراخوانی می‌شود.
</p>

<img src="img/Dynamic.png" alt="Dynamic Binding" />
</div>
<hr class="divider" />

```python
/*مقیدسازی ایستا*/

class Dog{
    private void eat() {
        System.out.println("dog is eating...");
    }

    public static void main(String[] args){
        Dog d1 = new Dog();
        d1.eat();
    }
}
```

```python
/*مقیدسازی پویا*/

class Animal{
    void eat() {
        System.out.println("animal is eating...")
    }
}

class Dog extends Animal{
    private void eat() {
        System.out.println("dog is eating...");
    }

    public static void main(String[] args){
        Animal a = new Dog();
        a.eat();
    }
}
```

<h3>تبدیل اشیا (Casting Objects)</h3>

<p>
در جلسات اول گفتیم که عملگر casting برای تبدیل متغیرهای یک نوع اصلی به نوع اصلی دیگر استفاده می‌شود.
</p>
<p>
این عملگر را می‌توان برای تبدیل یک شی از یک نوع کلاس به شی‌ای از کلاسی دیگر از طریق سلسله مراتب وراثتی مورد استفاده قرار داد.
</p>
<p>
متد زیر را در نظر بگیرید.
</p>
<p style="text-align: left;"><code> public static void m(Object x){
 System.out.println(x.toString());
 } </code>
</p>
<p>
در این متد یک متغیر ارجاعی از کلاس Object به عنوان پارامتر به متد ارسال می‌شود.
</p>

<p>
دستور زیر شی ()new Student را از نوع کلاس پدر جدش (Object) تعریف می‌کند.
</p>
<p style="text-align: left;"><code> ;()Object o = new Student </code></p>
<p>
دستور بالا یک تبدیل رو به بالا Upward casting نامیده می‌شود و تبدیلی ضمنی (Implicit casting) است. این تبدیل در جاوا مجاز است چون نمونه از کلاس Student به طور خودکار یک نمونه از کلاس Object به حساب می‌‌آید. 
</p>

<p>
سوالی مهم پیش می‌آید: <strong>چرا تبدیل لازم است؟</strong>
</p>

<p>
فرض کنید می‌خواهید ارجاع به شی o از کلاس Object را به متغیری از نوع Student با استفاده از دستور زیر انتساب دهید. 
</p>

<p>
با این دستور، یک خطای کامپایلری رخ خواهد داد. سوال اینجاست: چرا دستور <code>()Object o =new Student</code> اجرا می‌شود اما <code>Student b = o</code> کامپایل نمی‌شود؟
</p>
<p>
دلیل این است که شی Student همیشه یک نمونه از کلاس Object به حساب می‌آید، اما یک شی از Object لزوما یک نمونه از Student نیست. حتی اگر از نظر شما o واقعا شی‌ای از Student باشد، کامپایلر آنقدر باهوش نیست که این مسئله را درک کند.
</p>
<p>
برای اینکه به کامپایلر بگوییم o یک شی از Student است، از تبدیل آشکار (explicit) استفاده می‌کنیم. قاعده مورد استفاده، مشابه قاعده تبدیل نوع اصلی می‌باشد (استفاده از دو پرانتز برای نوع شی‌ای که می‌خواهیم تبدیل را بر روی آن انجام دهیم). 
</p>
<p style="text-align: left;"><code>Student b = (Student)o; //Explicit casting</code></p>

```python
/* مثالی از کاربرد مهم عملگر instanceof */

interface printable{}

class A implements printable{
    public void a() {
        System.out.println("a method");
    }
}
class B implements printable{
    public void b() {
        System.out.println("b method");
    }
}

class Call{
    void invoke(printable p){//upcasting
        if(p instanceof A){
            A a = (A)p; //downcasting
            a.a();
        }
        if(p instanceof B){
            B b = (B)p; //downcasting
            b.b();
        }
    }
}

class Test4{
    public static void main(String[] args){
        printable p = new B();
        Call c = new Call();
        c.invoke(p);
    }
}
```

مشخص است که تبدیل همواره در هر دو جهت قابل انجام نمی‌باشد. اگر با دستور "()new Carnivorous" یک شی گوشت‌خوار ایجاد کرده باشید شی مورد نظر نمی‌تواند مستقیما به یک سگ یا یک گربه  تبدیل شود، چون در واقع هیچ یک از آنها نمی‌باشد!
</p>
<p>
برای مثال تبدیل زیر نادرست است.
</p>
<p style="text-align: left;"><code>;()
Carnivorous m = new Carnivorous 
</code>
</p>
<p style="text-align: left;"><code>;
Cat c = (Cat)m
</code>
</p>
<p>
این کد با پیام زمان اجرای java.lang.ClassCastEception رو به رو می‌شود چون به زور می‌خواهیم گوشت‌خواری را که لزوما گربه نیست، به گربه تبدیل کنیم.
</p>
<p>
در حالت کلی برای اینکه بدانیم  چه موقع از تبدیل مناسب استفاده کنیم، این سوال را از خود بپرسیم : آیا گربه یک گوشت‌خوار است؟ بله، هست. پس تبدیل گربه به گوشت‌خوار قابل انجام است! 
</p>
<p>
دوباره بپرسیم: آیا یک گوشت‌خوار یک گربه است؟ پاسخ، خیر است. پس تبدیل گوشت‌خوار به گربه نمی‌تواند انجام بشود.
