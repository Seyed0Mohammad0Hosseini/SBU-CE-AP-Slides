  <h2>RandomAccessFile</h2>

  <p>
    این کلاس به ما این امکان را میدهد که درون فایل حرکت کرده و آن را پیمایش کنیم و بتوانیم از آن بخوانیم یا در آن بنویسیم.
  </p>

  <p>
  همچنین با استفاده از <strong>FileInputStream</strong> و <strong>FileOutputStream</strong> میتوان بخش هایی از فایل را با بخش هایی جدید جایگزین کرد.
  </p>


  <div class="slide-note">
    <strong>
    نکته :
    </strong>
    در جریان (stream) بر خلاف آرایه، اندیس برای خواندن و نوشتن معنی ندارد! و معمولا نمیتوان روی جریان به عقب و جلو حرکت کرد در حالی که در آرایه . RandomAccessFile امکان پذیر است.
    <br>
     اگرچه برخی از پیاده سازی های زیرکلاس stream مانند <strong>pushbackInputStream</strong> به ما این امکان را میدهد داده هارا در جریان push back کنیم تا بعدا دوباره بخوانیم، اما بخش محدودی از داده ها قابل push back هستند و نمیتوان پیمایش کاملی روی تمام داده ها داشت.
  </div>


  <hr class="divider" />

  <h3>
نحوه ایجاد یک RandomAccessFile :
  </h3>



</div>


```java
public class Test {
    public static void main(String[] args) {
        RandomAccessFile file = new RandomAccessFile("c:\\data\\file.txt","rw");
    }
}
```

    همچنین به یاد داشته باشید که این متد پس از عمل خواندن، اشاره گر فایل را درون فایل مربوطه<b> یک بایت به جلو میبرد</b>. یعنی نیاز نیست به صورت دستی اشاره گر را به جلو ببرید.

  <h3>
    نوشتن در فایل RandomAccessFile :
  </h3>
  <p>

  <h2>
    کلاس های InputStream و OutputStream
  </h2>
  <br>
  <p>
    این دو کلاس مبتنی بر بایت هستند، یعنی این جریان ها در هر لحظه یک بایت را میخوانند یا مینویسند.
  </p>
  <div class="slide-note">
    <strong>
    نکته :
    </strong>
      البته زیرکلاس های <strong>DataInputStream</strong> و <strong>DataOutputStream</strong> دو استثنا هستند که میتوانند در هر لحظه یک مقدار int, long, float و double بخوانند.
  </div>
  <br>
  <hr class="divider" />
  <br> 
  <h3>
  کلاس InputStream
  </h3>

  <p>
  این کلاس یک کلاس پایه و والد (و همچنین انتزاعی) برای تمام جریان های ورودی در جاوا میباشد، و اغلب برای خواندن داده ها از یک منبع داده ای از این کلاس استفاده میشود.
  <br>
  برای خواندن داده از فرزندان این کلاس، از متد read() استفاده میشود، این متد یک مقدار int برمیگرداند که حاوی مقدار بایت خوانده شده است.
  </p>
  <div class="slide-note">
    <strong>
    نکته :
    </strong>
    اگر داده ای برای خواندن باقی نمانده باشد، مقدار 1- را برمیگرداند.
  </div>
  <br>
  <h3>
  مثال :
  </h3>

</div>



```java
public class Test {
    public static void main(String[] args) {
        InputStream input = new FileInputStream("c:\\data\\input-file.txt");
        int data = input.read();
        while‌( data != -1 )
        {
            data = input.read();
        }
    }
}
```

<link rel="stylesheet" href="style.css">
<div class="card" dir="rtl">

  <p>
    همانطور که دیدید، <strong>FileInputStream</strong> یک از فرزندان کلاس انتزاعی <strong>java Inputstream</strong> میباشد. جاوا برای خواندن داده های بایتی از فایل ها از این کلاس استفاده میکند.
  </p>
  <h3>
  به مثال زیر توجه کنید:
  </h3>

</div>



```java
public class Test {
    public static void main(String[] args) {
        InputStream inputstream = new FileInputStream("c:\\data\\input-file.txt");
        int data = inputstream.read();
        while‌( data != -1 ){
            //do something with data .....
            data = inputstream.read();
        }
        inputstream.close();
    }
}
```

<link rel="stylesheet" href="style.css">
<div class="card" dir="rtl">

  <p>
    مثال بالا یک یک شیئ جدید از <strong>FileInputStream</strong> ایجاد میکند و متد ()read یک مقدار ظint برمیگرداند که حاوی مقدار Byte خوانده شده است که به صورت زیر میتوان آن را به char تبدیل کرد :
  </p>
  <p dir ='ltr'>
  char aChar = (char) data;
  </p>
  <p>
    اگر به پایان فایل رسیده باشیم و دیتایی برای خواندن باقی نمانده باشد، متد ()read مقدار  1- برمیگرداند.
  </p>
  <div class="slide-note">
    <strong>
    نکته :
    </strong>
       زیرکلاس های inputstream ممکن است دارای متد های ()read مختلفی باشند.
       <br>
      به عنوان مثال این متد در کلاس<strong>DataInputStream</strong> به شما این امکان را میدهد در هر لحظه یک مقدار int, long, float و double با متد های ()readDouble و ()readBoolean و ... بخوانید.
  </div>
  <br>
  <p>
    همچنین در کلاس <strong>InputStream</strong> دو متد ()read دیگر هم وجود دارد که میتواند داده ها را از منبع داده ای inputstream به درون آرایه از نوع بایت بریزد، خواندن یک آرایه از بایت ها در هر لحظه به مراتب سریع تر ازخواندن بایت  بایت داده ها میباشد.
    <br>
    این متد ها به صورت زیر هستند :
  </p>
  <p dir ='ltr'>
    int read(byte[]);<br>
    int read(byte[] , int offset , int length)
  </p>
  <br>
  <p>
    متد 
    int read(byte[] , int offset , int length)
    مانند قبلی یک آرایه از بایت هارا میخواند که از بایت offset شروع میشود و به طول length در آرایه قرار میدهد.
    متد read(byte[]) تلاش میکند تا حداکثر امکان بایت های بیشتری را بخواند و در ارایه پاس داده شده ذخیره کند.
    <br>
    این متد یک مقدار int برمیگرداند که میگویند چند Byte خوانده شده است، در مواقعی که تعداد Byte خوانده شده از inputstream کمتر از اندازه واقعی آرایه باشد، مقادیر سایر خانه ها تغییر نخواهد کرد و همان مقادیر قبلی باقی میماند.
  </p>

  <div class="slide-note">
    <strong>
    نکته :
    </strong>
     هر دو متد زمانی که به انتهای جریان میرسیم، مقدار 1- برمیگردانند.
  </div>
  <h3>
  مثال
  </h3>


</div>



```java
public class Test {
    public static void main(String[] args) {
        InputStream inputstream = new FileInputStream("c:\\data\\input-text.txt");
        byte[] data = new byte[1024];
        int bytesRead = inputstream.read(data);
        while(bytesRead != -1) {
            for(int i=0;i<bytesRead;i++) System.out.print(" "+data[i]);
            bytesRead = inputstream.read(data);
        }
        inputstream.close();
    }
}
```

  برای نوشتن داده ها در اشیای ساخته شده از <b>فرزندان </b> این کلاس از متد <strong>Write</strong> استفاده میشود.

<div class="card" dir="rtl">
  <h3>
  OutputStream و مقاصد داده ای
  </h3>

  <p>
  این کلاس هم یک کلاس پایه و والد (و همچنین انتزاعی) برای تمام جریان های خروجی در جاوا میباشد. برای نوشتن داده ها در یک مقصد داده ای اغلب از این کلاس استفاده میشود.
  <br>
    یک OutputStream به مقصد داده ای نظیر فایل، پایپ یا اتصال شبکه ای وصل میشود.
    <br>
    مقصد داده ای یک شیئ از کلاس OutputStream است که داده نوشته شده در شیئ سرانجام به آنجا منتقل میشود.
  </p>
  <br>


  <h3>
  متد Write(Byte) :
  </h3>
  <p>
    این متد برای نوشتن یک بایت در OutputStream میباشد، همچنین زیر کلاس های OutputStream دارای متد های ()Write دیگر هم هستند به عنوان مثال DataOutputStream به شما این امکان را میدهد داده هایی از نوع int, float, Double, Boolean, long و غیره را در قالب متد های ()writeBoolean یا ()writeDouble در یک مقصد داده ای بنویسید.
  </p>

  <hr class="divider" />
  <h3>
  مثال :
  </h3>
</div>



```java
public class Test {
    public static void main(String[] args) {
        OutputStream output = new FileOutputStream("c:\\data\\output-file.txt");
        while(hasMoreData()){
            int data = getMoreData();
            output.write(data);
        }
        output.close();
    }
}
```

<link rel="stylesheet" href="style.css">
<div class="card" dir="rtl">

  <p>
    در ابتدا یک شیئ از FileOutputStream ساخته میشود تا داده ها درون آن نوشته شوند، سپس در یک حلقه while تا زمانی که داده ای برای نوشتن وجود دارد، داده های در جریان نوشته میشوند.
    <br>
    درواقع در حلقه while داده از جای دیگر گرفته میشود و درون output ریخته میشود.
    <br>
    سر انجام با بستن جریان، محتوای آن به مقصد داده ای که در این مثال یک فایل است، منتقل خواهد شد.
    <br>
  </p>

  <hr class="divider" />

  <h3>
  متد Write(Byte[]) :
  </h3>
  <p>
    کلاس OutputStream دارای دو متد Write دیگر هم هست که میتوانند آرایه ای از نوع بایت(یا بخشی از آرایه را) به درون شیئ OutputStream بریزد.
    <br>
    متد ها به صورت زیر هستند‌:
    <br>
  </p>
  <p dir ='ltr'>
    write(byte[] bytes);<br>
    write(byte[] bytes , int offset , int length)
  </p>
  <p>
  متد اول، تمام بایت هارا به درون outputStream میریزد.
  <br>
  متد دوم تعداد length بایت را از اندیس offset آرایه bytes به درون outputStream میریزد
  </p>

  <hr class="divider" />

  <h3>
  متد ()flush :
  </h3>
  <p>
  این متد در کلاس outputStream ، تمام داده های قرار داده شده در شیئ از نوع outputStream را به مقصد داده ای منتقل میکند.
  <br>
  به عنوان مثال اگر outputStream یک FileOutputStream باشد تا قبل از فراخوانی متد ()flush داده های نوشته شده در آن به دیسک منتقل نمیشوند، بلکه در بخشی از حافظه بافر میشوند.
  <br>
  با فراخوانی متد flush اطمینان حاصل میکنیم که همه داده های بافر شده در مقصد نوشته خواهند شد.
  </p>
