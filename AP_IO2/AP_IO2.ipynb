  <h2>فایل و جریان</h2>



  <ul>
    <li> 
    .برای ذخیره کردن دیتا، آنها را در فایل ذخیره میکنیم که معمولا در حافظه جانبی ذخیره میشود.
    <br>
    چهار عمل اصلی CRUD روی فایل ها انجام میشود:
      <ul>
        <li><strong>Create : </strong>ایجاد کردن و ساخت فایل</li>
        <li><strong>Read‌ : </strong>خواندن دیتا از فایل</li>
        <li><strong>Update :‌ </strong>تغییر دیتای فایل</li>
        <li><strong>Delete‌ : </strong>حذف کردن فایل</li>
      </ul>
    </li>
    <li>
        انواع ذخیره سازی فایل :
        <ul>
          <li><strong>متنی (text files) : </strong>از کاراکتر ها تشکیل شده اند، مانند فایل های txt و html </li>
          <li><strong>باینری (binary files) : </strong>مانند فایل های zip,pdf,exe</li>
      </ul>
    </li>
  </ul>
  

  <p>
  <strong>
  Stream چیست؟
  </strong>
  </p>

  <p>
  یک جریان از داده ها میباشد که مثل خط لوله داده میتواند از طریق آن جریان یابد، درواقع یک دنباله پیوسته از داده هاست.
  <br>
  در برنامه میتواند ورودی یا خروجی باشد،همچنین جریان داده میتواند برای داده های متنی، یا داده های باینری باشد.
  <br>
  به عنوان مثال در تصویر زیر، جریان داده سمت چپ ورودی برنامه ما، و جریان داده سمت راست خروجی برنامه است.
  <br>
  و همانطو که مشخص است هم میتواند به فایل متصل شود یا به پایپ یا به شبکه متصل شود.
  </p>

  <p>
  <br>

  <p>

<div class="card" dir="rtl">
  <h2>RandomAccessFile</h2>

  <p>
    این کلاس به ما این امکان را میدهد که درون فایل حرکت کرده و آن را پیمایش کنیم و بتوانیم از آن بخوانیم یا در آن بنویسیم.
  </p>

  <p>
  همچنین با استفاده از <strong>FileInputStream</strong> و <strong>FileOutputStream</strong> میتوان بخش هایی از فایل را با بخش هایی جدید جایگزین کرد.
  </p>


  <div class="slide-note">
    <strong>
    نکته :
    </strong>
    در جریان (stream) بر خلاف آرایه، اندیس برای خواندن و نوشتن معنی ندارد! و معمولا نمیتوان روی جریان به عقب و جلو حرکت کرد در حالی که در آرایه . RandomAccessFile امکان پذیر است.
    <br>
     اگرچه برخی از پیاده سازی های زیرکلاس stream مانند <strong>pushbackInputStream</strong> به ما این امکان را میدهد داده هارا در جریان push back کنیم تا بعدا دوباره بخوانیم، اما بخش محدودی از داده ها قابل push back هستند و نمیتوان پیمایش کاملی روی تمام داده ها داشت.
  </div>


  <hr class="divider" />

  <h3>
نحوه ایجاد یک RandomAccessFile :
  </h3>



</div>


```java
public class Test {
    public static void main(String[] args) {
        RandomAccessFile file = new RandomAccessFile("c:\\data\\file.txt","rw");
    }
}
```

<link rel="stylesheet" href="style.css">
<div class="card" dir="rtl">


  

  <p>
    پارامتر دوم (rw) مُد باز کردن فایل را مشخص میکند، و "rw" به معنای مد خواندن/نوشتن است.
  </p>
  <hr class="divider" />
  <p>
    برای خواندن و نوشتن در محل خاص از یک RandomAccessFile، باید ابتدا اشاره گر فایل را در محل مورد نظر قرار دهیم.
    <br>
    این کار را با تابع seek() انجام میدهیم، همچنین با تابع getFilePointer() از محل فعلی اشاره گر آگاه میشویم :
  </p>


</div>


```java
public class Test {
    public static void main(String[] args) {
        RandomAccessFile file = new RandomAccessFile("c:\\data\\file.txt","rw");
        long pointer = file.getFilePointer();
        file.close();
    }
}
```

  </div>

  <br>

  <hr class="divider" />

  <h3>
    خواندن از RandomAccessFile :
  </h3>
  <p>
    برای خواندن از یک RandomAccessFile میتوان یکی از متد های read() را انتخاب کرد، به عنوان مثال :
  </p>


</div>



```java
public class Test {
    public static void main(String[] args) {
        RandomAccessFile file = new RandomAccessFile("c:\\data\\file.txt","rw");
        int aByte = file.read();
        file.close();
    }
}
```

<link rel="stylesheet" href="style.css">
<div class="card" dir="rtl">

  <p>
    متد read() بایتی را که در محل اشاره گر فایل در شیئ ایجاد شده اس <strong>RandomAccessFile</strong> قرار دارد، میخواند.
    <br>
    همچنین به یاد داشته باشید که این متد پس از عمل خواندن، اشاره گر فایل را درون فایل مربوطه<b> یک بایت به جلو میبرد</b>. یعنی نیاز نیست به صورت دستی اشاره گر را به جلو ببرید.
  </p>

  <br>
  <hr class="divider" />
  <br>
  <h3>
    نوشتن در فایل RandomAccessFile :
  </h3>
  <p>
  برای نوشتن در این نوع فایل ها هم متد های مختلفی وجود دارد که میتوان از آنها استفاده کرد، به عنوان مثال میتوان از متد write()  استفاده کرد که آن هم مانند متد read() پس از اجرا، اشاره گر را به یک بایت جلوتر حرکت میدهد :
  </p>

</div>



```java
public class Test {
    public static void main(String[] args) {
        RandomAccessFile file = new RandomAccessFile("c:\\data\\file.txt","rw");
        file.write("Hello World".getBytes());
        file.close();
    }
}
```
