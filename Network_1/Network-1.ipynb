  <h2>آدرس IP</h2>
  <ul>
    <li>آدرس IP یک عدد یکتاست که به هر گره (node) در شبکه اختصاص داده می‌شود؛ مثل <code dir="ltr">192.168.0.1</code>.

      <br>هر اکتت در واقع یک بایت است و جمع آن‌ها 32 بیت را می‌سازد.</li>
    <li>یک آدرس <strong>منطقی</strong> است و می‌تواند تغییر کند.

<div class="card" dir="rtl">
  <h2>پروتکل (Protocol)</h2>
  <p>
    پروتکل یعنی «قواعد گفت‌وگو» بین دستگاه‌ها: پیام چه شکلی باشد، از کدام پورت برود و ارتباط کی شروع و تمام شود؛
    اگر این قواعد نباشند، دو طرف حتی با داشتن شبکه مشترک هم عملا حرف یکدیگر را نمی‌فهمند.
  </p>
  <ul>
    <li><strong>TCP</strong>: این پروتکل ابتدا ارتباط را ایجاد و پایدار می‌کند (handshake سه‌مرحله‌ای) و بعد داده را به قطعات کوچک تقسیم می‌کند؛
      برای هر قطعه رسید می‌گیرد، در صورت گم‌شدن دوباره می‌فرستد و با شماره‌گذاری مطمئن می‌شود همه بسته‌ها دقیقا به همان ترتیبی که ارسال شده‌اند
      به مقصد می‌رسند. TCP در کنار <strong>IP</strong> کار می‌کند تا مسیر درست انتخاب شود و به‌خاطر اطمینان بالا، برای وب و ایمیل و هر جایی که
      «درست رسیدن» مهم‌تر از «سریع رسیدن» است، انتخاب رایجی به‌شمار می‌آید.</li>
    <li><strong>UDP</strong>: در UDP هیچ مذاکره اولیه‌ای برای ساختن ارتباط انجام نمی‌شود و فرستنده بسته‌ها را مستقیم راهی می‌کند؛ خبری از رسید و ارسال‌مجدد
      یا تضمین ترتیب نیست و همین سادگی باعث می‌شود سربار بسیار کم و تاخیر پایین باشد. به همین دلیل، برای پخش زنده، تماس صوتی/تصویری و بازی‌های آنلاین
      که «به‌موقع رسیدن» مهم‌تر از «بی‌نقص بودن» است، معمولا مناسب‌تر از TCP است.</li>
    <li><strong>FTP</strong>: پروتکلی برای انتقال فایل که امکاناتی مثل فهرست‌کردن، ساخت پوشه، ازسرگیری دانلود/آپلود و جابه‌جایی فایل بین کلاینت و سرور را فراهم می‌کند؛
      در سناریوهای حساس معمولا از گونه‌های امن آن مثل <strong>SFTP</strong> یا <strong>FTPS</strong> استفاده می‌شود تا هم محتوا و هم نام کاربری/گذرواژه رمزنگاری شوند.</li>
    <li><strong>Telnet</strong>: راهی قدیمی برای دسترسی به خط فرمان سیستم‌های راه‌دور است و اجازه می‌دهد مثل کنسول محلی دستور اجرا کنید؛ اما چون هیچ رمزنگاری‌ای
      انجام نمی‌دهد، شنود اطلاعات بسیار ساده است و به همین خاطر امروز تقریبا همیشه <strong>SSH</strong> به‌عنوان جایگزین امن توصیه می‌شود.</li>
    <li><strong>SMTP</strong>: ستون فقرات ارسال ایمیل در اینترنت است؛ کلاینت‌ها پیام را از طریق SMTP به سرور می‌سپارند و سرورها نیز بین خودشان پیام را جابه‌جا می‌کنند
      تا به صندوق گیرنده برسد. پورت‌های رایج 25 و 587 و 465 هستند که حالت‌های امن با <strong>TLS/SSL</strong> را هم پوشش می‌دهند.</li>
    <li><strong>POP</strong>: پروتکل دریافت ایمیل که معمولا پیام‌ها را از سرور دانلود می‌کند و روی دستگاه کاربر نگه می‌دارد؛ این روش برای استفاده تک‌دستگاه ساده و مناسب است،
      اما اگر همگام‌سازی پوشه‌ها روی چند دستگاه را می‌خواهید، <strong>IMAP</strong> تجربه بهتری می‌دهد. نسخه <em>POP3</em> روی پورت‌های 110 و 995 (نسخه امن) کار می‌کند.</li>
  </ul>

<div class="card" dir="rtl">
  <h2>شماره پورت (Port Number)</h2>
  <ul>
    <li>شماره پورت برای <strong>تشخیص یکتای برنامه‌ها</strong> روی یک میزبان (Host) استفاده می‌شود؛ چند برنامه می‌توانند روی یک IP باشند ولی هرکدام پورت خودشان را دارند.</li>
    <li>پورت در عمل <strong>نقطه انتهایی ارتباط</strong> بین دو برنامه است؛ داده‌ها به «IP + پورت» مقصد ارسال می‌شوند و به برنامه درست تحویل داده می‌شوند.</li>
    <li>برای ارتباط، معمولا ترکیب <strong>آدرس IP</strong> به‌همراه <strong>شماره پورت</strong> مبدا/مقصد مشخص می‌شود؛ مثلا <code dir="ltr">192.168.1.10:8080</code>.</li>
  </ul>
  <hr class="divider" />

  <h2>روش‌های مختلف اتصال</h2>
  <ul>
    <li>در <strong>پروتکل‌های اتصال‌گرا</strong> (مثل TCP) گیرنده برای بسته‌های دریافت‌شده <strong>تأییدیه (ACK)</strong> می‌فرستد؛
      بنابراین ارتباط <strong>قابل اعتماد</strong> است اما به‌خاطر کنترل‌ها و رفت‌وبرگشت‌های بیشتر، معمولا <strong>کندتر</strong> از حالت بی‌اتصال عمل می‌کند.</li>
    <li>در <strong>پروتکل‌های بی‌اتصال</strong> (مثل UDP) گیرنده تأییدیه‌ای نمی‌فرستد؛ پس ارتباط <strong>سریع</strong> است اما
      <strong>تضمین تحویل و ترتیب</strong> ندارد و ممکن است بسته‌ها گم شوند یا جابه‌جا برسند.</li>
  </ul>

<div class="card" dir="rtl">
  <h2>متدها و سازنده‌های کلاس ServerSocket</h2>
  <ul>
    <li>کلاس <code>java.net.ServerSocket</code> برای برنامه‌های <strong>سرور</strong> به‌کار می‌رود تا روی یک <strong>پورت</strong> خاص گوش بدهند و درخواست‌های کلاینت را بپذیرند.</li>
    <li>این کلاس <strong>چهار سازنده</strong> دارد. اگر در سازنده <strong>exception</strong> رخ ندهد، یعنی برنامه با موفقیت به پورت مشخص‌شده <strong>bind</strong> شده و آماده دریافت اتصال است.</li>
  </ul>
  <table>
    <tbody>
      <tr>
        <td>1)</td>
        <td><code>public ServerSocket(int port)</code></td>
        <td>تلاش می‌کند سرور سوکتی بسازد که به پورت داده‌شده متصل است. اگر آن پورت قبلا توسط برنامه دیگری اشغال شده باشد، <code>IOException</code> رخ می‌دهد.</td>
      </tr>
      <tr>
        <td>2)</td>
        <td><code>public ServerSocket(int port, int backlog)</code></td>
        <td>مثل سازنده قبلی، با این تفاوت که پارامتر <strong>backlog</strong> مشخص می‌کند چند اتصال ورودی در حال انتظار در صف نگه‌داری شوند.</td>
      </tr>
      <tr>
        <td>3)</td>
        <td><code>public ServerSocket(int port, int backlog, InetAddress address)</code></td>
        <td>به‌جز تعیین پورت و صف انتظار، با پارامتر <code>InetAddress</code> مشخص می‌کنید سرور روی کدام <strong>IP محلی</strong> bind شود (برای سیستم‌هایی که چند IP دارند).</td>
      </tr>
      <tr>
        <td>4)</td>
        <td><code>public ServerSocket()</code></td>
        <td>یک سرورسوکتِ <strong>غیرمتصل</strong> می‌سازد؛ بعدا با متد <code>bind()</code> آن را به IP/Port موردنظر متصل کنید.</td>
      </tr>
    </tbody>
  </table>
  <br>
  <h3>راه‌اندازی ساده Client/Server</h3>
  <ul>
    <li>در سناریوی ساده یک‌طرفه، کلاینت وصل می‌شود و پیام می‌فرستد و سرور پیام‌ها را روی اتصال سوکت نمایش می‌دهد. جزئیات سطح پایین را بسته شبکه جاوا (<code>java.net</code>) مدیریت می‌کند و کار برنامه‌نویس آسان‌تر می‌شود..</li>
    <li>برای اتصال به یک نود دیگر، باید یک <strong>اتصال سوکت</strong> برقرار شود؛ یعنی دو نود از <strong>آدرس IP</strong> و <strong>پورت TCP</strong> یکدیگر اطلاع داشته باشند. کلاس <code>java.net.Socket</code> نماینده یک سوکت در سمت کلاینت است.</li>
  </ul>
  <div class="code-block">
<pre><code dir="ltr">Socket socket = new Socket("127.0.0.1", 6666);
</code></pre>
  </div>
  <ul>
    <li><strong>Arg 1:</strong> آدرس IP سرور. مقدار <code>127.0.0.1</code> همان <em>localhost</em> است و وقتی کلاینت و سرور روی یک دستگاه هستند به کار می‌آید.</li>
    <li><strong>Arg 2:</strong> شماره پورت TCP. فقط یک عدد برای مشخص‌کردن برنامه مقصد؛ مثلا HTTP روی پورت <code>80</code> است. بازه پورت‌ها از <code>0</code> تا <code>65535</code> است.</li>
  </ul>

  <h2>ایجاد سرور (Server)</h2>

  <h2>ارتباط روی سوکت و بستن اتصال</h2>
  <ul>
    <li><strong>Communication</strong>: برای تبادل داده روی اتصال سوکت از <strong>Stream</strong> استفاده می‌کنیم: <code>getInputStream()</code> و <code>getOutputStream()</code>.
      برای کار راحت‌تر از <code>BufferedReader/PrintWriter</code> یا <code>DataInputStream/DataOutputStream</code> نیز استفاده می‌کنیم.</li>
    <li><strong>Establish a Socket Connection</strong>: در یک برنامه سرور، عملا با <strong>دو سوکت</strong> سروکار داریم:
      <ul>
        <li>یک <code>ServerSocket</code> که منتظر اتصال کلاینت‌ها می‌ماند (متد <code>accept()</code>).</li>
        <li>یک <code>Socket</code> که نتیجه <code>accept()</code> است و برای ارسال/دریافت داده با همان کلاینت استفاده می‌شود.</li>
      </ul>
    </li>
    <li><strong>Closing the connection</strong>: پس از اتمام کار، جریان‌های ورودی/خروجی و سپس خودِ <code>Socket</code> را ببندید تا منابع آزاد شوند؛
      استفاده از <code>try-with-resources</code> جلوی هدررفت منابع را خواهد گرفت.</li>
  </ul>
  <hr class="divider" />
</div>



```java
import java.io.*;
import java.net.*;

public class MyServer {
    public static void main(String[] args) {
        try {
            ServerSocket ss = new ServerSocket(6666);
            Socket s = ss.accept(); // establishes connection

            DataInputStream dis = new DataInputStream(s.getInputStream());
            String str = dis.readUTF();
            System.out.println("message= " + str);

            dis.close();
            s.close();
            ss.close();
        } catch (Exception e) {
            System.out.println(e);
        }
    }
}

```

```java
import java.io.*;
import java.net.*;

public class MyClient {
    public static void main(String[] args) {
        try {
            Socket s = new Socket("localhost", 6666);

            DataOutputStream dout = new DataOutputStream(s.getOutputStream());
            dout.writeUTF("Hello Server");
            dout.flush();

            dout.close();
            s.close();
        } catch (Exception e) {
            System.out.println(e);
        }
    }
}

```

  <h2>مثال سوکت در جاوا (خواندن و نوشتن دوطرفه)</h2>
  <ul>
    <li>در این مثال، <strong>کلاینت</strong> ابتدا به سرور پیام می‌فرستد و <strong>سرور</strong> آن را دریافت و چاپ می‌کند.</li>
    <li>سپس <strong>سرور</strong> پاسخی برای کلاینت می‌نویسد و <strong>کلاینت</strong> آن را دریافت و چاپ می‌کند.</li>
    <li>این روند به صورت رفت‌وبرگشتی ادامه می‌یابد تا زمانی که یکی از طرفین عبارت <code dir="ltr">stop</code> را ارسال کند و ارتباط خاتمه یابد.</li>
  </ul>
  <hr class="divider" />
</div>

```java
import java.net.*;
import java.io.*;

class MyServer {
    public static void main(String args[]) throws Exception {
        ServerSocket ss = new ServerSocket(3333);
        Socket s = ss.accept();

        DataInputStream din = new DataInputStream(s.getInputStream());
        DataOutputStream dout = new DataOutputStream(s.getOutputStream());
        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));

        String str = "", str2 = "";
        while (!str.equals("stop")) {
            str = din.readUTF();
            System.out.println("client says: " + str);
            str2 = br.readLine();
            dout.writeUTF(str2);
            dout.flush();
        }

        din.close();
        s.close();
        ss.close();
    }
}

```

```java
import java.net.*;
import java.io.*;

class MyClient {
    public static void main(String args[]) throws Exception {
        Socket s = new Socket("localhost", 3333);

        DataInputStream din = new DataInputStream(s.getInputStream());
        DataOutputStream dout = new DataOutputStream(s.getOutputStream());
        BufferedReader br = new BufferedReader(new InputStreamReader(System.in));

        String str = "", str2 = "";
        while (!str.equals("stop")) {
            str = br.readLine();
            dout.writeUTF(str);
            dout.flush();
            str2 = din.readUTF();
            System.out.println("Server says: " + str2);
        }

        dout.close();
        s.close();
    }
}

```
